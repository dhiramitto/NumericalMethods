# Runge-Kutta Method 3rd and 4th Order

```python
import numpy as np
def f(x,y):
    return x**2-4*y

def fex(x):
    return (31/32)*np.exp(-4*x)+(1/4)*x**2-(1/8)*x+(1/32)

def runge_kutta3(a,b,x,y,n):
    h = (b-a)/(n-1)
    ya = np.zeros(5)

    for i in range(5):
        k1 = f(x,y)
        k2 = f(x+0.5*h,y+0.5*h*k1)
        k3 = f(x+h,y-k1*h+2*k2*h)
        ya[i] = y
        print('%d. %.2f %.3f %.3f %.3f %.3f' % ((i+1), x, y, k1, k2, k3))
        y = y + h*(k1+4*k2+k3)/6
        x = x + h
    return ya

y3 = runge_kutta3(0,1,0,1,5)
```

```python
def runge_kutta4(a,b,x,y,n):
    h = (b-a)/(n-1)
    ya = np.zeros(5)

    for i in range(5):
        k1 = f(x,y)
        k2 = f(x+0.5*h,y+0.5*h*k1)
        k3 = f(x+0.5*h,y+0.5*k2*h)
        k4 = f(x+h,y+h*k3)
        ya[i] = y
        print('%d. %.2f %.3f %.3f %.3f %.3f %.3f' % ((i+1), x, y, k1, k2, k3, k4))
        y = y + h*(k1+2*k2+2*k3+k4)/6
        x = x + h
    return ya

y4 = runge_kutta4(0,1,0,1,5)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = np.linspace(0,1,5)

plt.figure(figsize=(18,7))
plt.plot(x,y3,color='red', label='Runge-Kutta Method 3rd Order')
plt.plot(x,y3,'or')
plt.plot(x,y4,color='green', label='Runge-Kutta Method 4th Order')
plt.plot(x,y4,'og')
plt.plot(x,fex(x),color='blue',label='Exact')
plt.plot(x,fex(x),'vb')
plt.legend()
plt.show()
```
