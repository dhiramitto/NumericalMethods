## B. Heat Conduction Equation

Diketahui soal sebagai berikut,

Use the explicit method to solve for the temperature distribution of a long, thin rod with a length of $10 \, cm$ and the following values: $\displaystyle k’ = 0.49 \,cal\,/\,(s\cdot cm \cdot ^{\circ}C)$, $\Delta x = 2 cm$, and $\Delta t = 0.1 \,s$. At $t = 0$, the temperature of the rod is zero and the boundary conditions are fixed for all times at $T(0) = 1000^{\circ}C$ and $T(10) = 500^{\circ}C$. Note that the rod is aluminum with $C = 0.2174 \, \, cal\,/\,(g\cdot ^{\circ}C)$ and $r = 2.7 g/cm^3$. Therefore, $k = 0.49/(2.7 \cdot 0.2174) = 0.835 cm^2/s$ and $\lambda = 0.835(0.1)/22 = 0.020875$.

Soal di atas, akan kita selesaikan menggunakan Python di bawah ini:

```python
import numpy as np
import matplotlib.pyplot as plt

T=np.zeros((121,6))

T[:,0]=100
T[:,5]=50
k=0.835
lamda=0.020875
dt=0.1
l=0
i=0

while(l<120):
    for i in range(1,5):
        T[l+1,i]=T[l,i]+lamda*(T[l,i+1]-2*T[l,i]+T[l,i-1])
    l=l+1;

print(T)
```

Solusi di atas, dapat divisualisasikan dalam grafik berikut:

```python
x=np.array([0,2,4,6,8,10])
y0=np.array(T[29,:])
y1=np.array(T[59,:])
y2=np.array(T[89,:])
y3=np.array(T[119,:])

plt.plot(x,y0,'r')
plt.plot(x,y1,'b')
plt.plot(x,y2,'y')
plt.plot(x,y3,'g')

plt.plot(x,y0,'ro')
plt.plot(x,y1,'bo')
plt.plot(x,y2,'yo')
plt.plot(x,y3,'go')

plt.text(4,50,'t=12')
plt.text(4,39,'t=9')
plt.text(4,28,'t=6')
plt.text(4,12,'t=3')

plt.show()
```
